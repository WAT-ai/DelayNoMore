```python
import river.drift
import numpy as np
```

```python
adwin = river.drift.ADWIN()
```

```python
stream_length = 1000
data_stream = np.random.normal(loc=0, scale=1, size=stream_length // 2)  # First distribution
data_stream = np.concatenate([data_stream, np.random.normal(loc=5, scale=1, size=stream_length // 2)])  # Second distribution (drift)
```

```python
drift_detected = []
for i, x in enumerate(data_stream):
    adwin.update(x)
    if adwin.drift_detected:
        drift_detected.append(i)  # Record the point of drift detection

print(data_stream)

print(f"Drift detected at indices: {drift_detected}")
```
