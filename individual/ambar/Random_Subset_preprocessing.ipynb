```python
import boto3
import pickle
import pandas as pd
from io import BytesIO
import sys
```

```python
bucket_name = 'delaynomore-data-pond'
file_key = 'data/random_subset_100K.pkl'

# Create an S3 client
s3 = boto3.client('s3')

# Download the file as an in-memory object
response = s3.get_object(Bucket=bucket_name, Key=file_key)
print("File Size (in bytes):", response['ContentLength'])
stream = response['Body']
```

```python
data = pickle.load(stream)
sys.getsizeof(data)
```
